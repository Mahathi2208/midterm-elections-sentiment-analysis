```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
df = pd.read_csv('twitter_scrapped_final.csv')
df
```

```python
word_freq_map = {}
stop_words = list(set(stopwords.words('english')))
words = set(nltk.corpus.words.words())
```

```python
# remove duplicate
tweet_list = list(set(df['text']))
```

```python
for comment in tweet_list:
    for word in comment.split():
        
        word = re.sub(r'[^\w]', ' ', word).strip().lower()
        
        if " " in word:
            for w in word.split():
                w = re.sub(r'[^\w]', ' ', w).strip().lower()

                if w in stop_words:
                    continue
            
                if w in word_freq_map:
                    word_freq_map[w] += 1
                else:
                    word_freq_map[w] = 1
        else:
            if word in stop_words:
                    continue
            if word in word_freq_map:
                word_freq_map[word] += 1
            else:
                word_freq_map[word] = 1
```

```python
word_freq_map
```

```python
new_list = sorted(word_freq_map.items(), key = lambda x: -x[1])
word_list = [w[0] for w in new_list]
cnt_list = [w[1] for w in new_list]

word_count_df = pd.DataFrame({'word': word_list, 'frequency': cnt_list})
word_count_df
```

```python
word_count_df.to_csv('word_cnt.csv')
```

```python
Fetterman_list = ['johnfetterman', 'fetterman', 'fettermanforsenate', 'votefetterman', 'fettermanisthebetterman','fettermanforpa', 'giselefetterman','levifetterman','fettermanlt',
                'fetterwoman','fetterman2022','fettermania','fettermans', 'fettermandailymeme','fetter','votejohnfetterman','fettermen' ]
Oz_list = ['mehmetoz','mehmet', 'mehmetozthepuppymurderer', 'mehmetfromjersey', 'defundmehmetoz','snakeoilmehmetöz', 'mehmetozkillspuppies',
            'oz','voteoz','drozkillspuppies','droz','ooze','rejectoz','dogkilleroz','nooz','sendozhome','stopoz','ozisafraud','notooz', 'noozinpa', 'noonoz', 'voz',
            'votedroz', 'neveroz', 'ozfornj', 'saynotooz', 'ozzie', 'ozzy', 'realdoctorsagainstoz', 'ozforsenate', 'carpetbaggeroz', 'drozhitlercar', 'ozkillspuppies',
            'drozisafraud', 'ozisaquack', 'Dr. Oz', 'Dr Oz']
```

```python
candidate_map = {}

for word in Fetterman_list:
    candidate_map[word] = "Fetterman"
for word in Oz_list:
    candidate_map[word] = "Oz"
```

```python
for i in range(len(comment)): 
    for word in Fetterman_list:
        if word in comment[i]:
            comment[i] = comment[i].replace(word, candidate_map[word])
        
    for word in Oz_list:
        if word in comment[i]:
            comment[i] = comment[i].replace(word, candidate_map[word])
```

```python
word_pair = []
candidate_list = ["Fetterman", "Oz"]

for line in comment:
    for candidate in candidate_list:
        if "Dr. Oz" in line:
            print(line)
        if candidate in line.split():
            line_list = line.split()
            new_line = []
            for word in line_list:
                if word in stop_words:
                    continue
                new_line.append(word)
            idx = new_line.index(candidate)
            if idx + 2 < len(new_line) and idx - 2 >= 0:
                word_pair.append((new_line[idx-2] + " " + new_line[idx-1] + " " + new_line[idx] + " " + new_line[idx+1] + " " +new_line[idx+2]))
            elif idx + 1 < len(new_line):
                word_pair.append((new_line[idx] + " " + new_line[idx+1]))
            elif idx - 1 >= 0:
                word_pair.append(( new_line[idx-1] + " " + new_line[idx]))
```
